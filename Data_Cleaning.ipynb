```python
title=pd.read_csv('/content/drive/Shared drives/CIS 550/title.basics.tsv',sep='\t',header=0)
#title = title.loc[title['runtimeMinutes'] != '\\N']
title=title.loc[title['titleType']=='movie']
title['primaryTitle']=title['primaryTitle'].str.strip()
title['originalTitle']=title['originalTitle'].str.strip()
title.head()
```

```python
title=title.drop(columns=['titleType','originalTitle','isAdult','runtimeMinutes','endYear','genres'])
title=title.rename(columns={'tconst':'movie_id'})
title.head()
```

```python
ratings=pd.read_csv('/content/drive/Shareddrives/CIS 550/title.ratings.tsv',sep='\t',header=0)
ratings=ratings.drop(columns=['numVotes'])
ratings=ratings.rename(columns={'tconst':'movie_id'})
ratings.head()
```

```python
movie_forsql=title.merge(ratings)
movie_forsql.head()
```

```python
len(movie_forsql)
```

```python
movie_forsql.to_csv("/content/drive/Shared drives/CIS 550/movie.csv",index=None)
```

```python
actor_name=pd.read_csv('/content/drive/Shared drives/CIS 550/name.basics.tsv',sep='\t',header=0)
actor_name['primaryName']=actor_name['primaryName'].str.strip()
actor_name.head()
```

```python
actor_name=actor_name.drop(columns=['birthYear','deathYear','primaryProfession','knownForTitles'])
actor_name=actor_name.rename(columns={'nconst':'actor_id'})
actor_name.head()
```

```python
cast=pd.read_csv('/content/drive/Shared drives/CIS 550/title.principals.tsv',sep='\t',header=0)
cast = cast.loc[cast['characters'] != '\\N']
cast['characters']=cast['characters'].str.replace(r'\[\"', '') 
cast['characters']=cast['characters'].str.replace(r'\"\]', '') 
cast['characters']=cast['characters'].str.strip()
cast.head()
```

```python
cast1=cast.drop(columns=['ordering','category','job'])
cast1.head()
```

```python
cast1.isnull().values.any()
```

```python
mbti=pd.read_csv('/content/drive/Shared drives/CIS 550/mbti.csv')
mbti=mbti.dropna(subset=['movie'])
mbti['movie']=mbti['movie'].str.replace(r"\(.*\)","")
mbti['movie']=mbti['movie'].str.strip()
mbti.head()
```

```python
mbti['role']=mbti['role'].str.strip()
mbti1=mbti.drop(columns=['stat','enneagram'])
```

```python
cast_withmovie=cast1.merge(title,left_on='tconst',right_on='movie_id',how='inner')
cast_withmovie.head()
```

```python
cast_withmovie1=cast_withmovie.drop(columns=['tconst','startYear'])
cast_withmovie1.head()
```

```python
cast_withmovie1[['characters','primaryTitle']].isnull().values.any()
```

```python
mbti1[['role','movie']].isnull().values.any()
```

```python
characters2=cast_withmovie1.merge(mbti1,left_on=['characters','primaryTitle'],right_on=['role','movie'],how='left')
characters1=characters2.drop(columns=['role','movie','primaryTitle'])
characters1.head()
```

```python
characters=characters1.drop(columns='nconst')
characters=characters.rename(columns={'characters':'Name'})
characters.head()
```

```python
characters.count()
```

```python
characters=characters.drop_duplicates(subset=['Name', 'movie_id'], keep='last')
```

```python
characters.count()
```

```python
characters.to_csv("/content/drive/Shared drives/CIS 550/characters.csv",index=None)
```
